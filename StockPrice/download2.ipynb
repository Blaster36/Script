```python
import datetime
import pandas_datareader.data as web
import csv
```

```python
start = datetime.date(2018, 1, 1)
end = datetime.date(2021, 1, 1)
```

```python
# ソフトバンクグループ（証券コード：9984）の株価データ取得
df = web.DataReader('9984.T', 'yahoo', start, end)
df.head()
```

```python
df.to_csv('9984.T.csv', encoding='utf-8', index=True, header=True, quotechar='"', sep=',', quoting=csv.QUOTE_ALL)
```

```python
# 日経225の株価データ取得
df = web.DataReader('^N225', 'yahoo', start, end)
df.head()
```

```python
df.to_csv('N225.csv', encoding='utf-8', index=True, header=True, quotechar='"', sep=',', quoting=csv.QUOTE_ALL)
```
