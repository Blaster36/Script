```python
# pandasのインポート
import pandas as pd
```

```python
# データの読み込み
df = pd.read_csv('finance_dataset.csv')
# データフレームの表示
df
```

```python
# 各カラムの詳細確認
df.info()
```

```python
# Dateのデータ型を'datetime'型へ変更
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
# 曜日情報を追加（月曜:0, 火曜:1, 水曜:2, 木曜:3, 金曜:4, 土曜:5, 日曜:6）
df['weekday'] = df['Date'].dt.weekday
df
```

```python
# 初めの月曜日となる1965/1/4を基準に週を追加
from datetime import datetime
from datetime import timedelta
start = datetime(1965, 1, 4)
df['weeks'] = (df['Date'] - start) // timedelta(weeks=1)
df
```

```python
# カラムの並び替え
df = df[['Date', 'weeks', 'weekday', 'High', 'Low', 'Open', 'Close', 'Volume']]
df
```

```python
# 日付をインデックスにセット
df.set_index(keys='Date', inplace=True)
df

# データの並び替え
df.sort_values(by='Date', ascending=True, inplace=True)
```

```python
# カラム情報を1行上にずらしたデータフレームを作成する
df_shift = df.shift(-1)
df_shift
```

```python
# 翌日の終値と本日の終値の差分を追加する
df['delta_Close'] = df_shift['Close'] - df['Close']
df
```

```python
# 目的変数Upを追加する（翌日の始値が上がる場合1、それ以外は0とする
df['Up'] = 0
df['Up'][df['delta_Close'] > 0] = 1
df = df.drop('delta_Close', axis=1)
df
```

# データの全体像をつかむ

```python
# 'Open', 'High', 'Low', 'Close'グラフ化のためにカラム抽出
df_new = df[['Open', 'High', 'Low', 'Close']]

# matplotlibのインポート
from matplotlib import pyplot as plt
%matplotlib inline

# 時系列折れ線グラフの作成
df_new.plot(kind='line')
```

# 特徴量を追加する

```python
# 終値の前日比の追加
df_shift = df.shift(1)

df['Close_ratio'] = (df['Close'] - df_shift['Close']) / df_shift['Close']
df
```

```python
# 始値と終値の差分を追加
df['Body'] = df['Open'] - df['Close']
df
```

# 不要なデータを削除する

```python
# 週番号をリストに格納
list_weeks = []
list_weeks = df['weeks'].unique()
list_weeks
```

```python
# 各週ごろの日数を入力
df['weeks_days'] = 0
for i in list_weeks:
    df['weeks_days'][df['weeks'] == i] = len(df[df['weeks'] == i])
df
```

```python
# 月曜～金曜まで5日分データのある週だけデータを取り出す
df = df[df['weeks_days'] == 5]
df
```

```python
# 金曜日のデータを削除する（weekday:4となるデータ
df = df[df['weekday'] != 4]
df
```

```python
# 不要カラムの削除と並べ替え
df = df[['weekday', 'High', 'Low', 'Open', 'Close', 'Close_ratio', 'Body', 'Up']]
df
```

```python
# 学習データを2018-01-01～2020-12-31の期間とし、df_trainに入力する
df_train = df['2018-01-01':'2020-12-31']
df_train
```

```python
# 検証データを説明変数（X_val）と目的変数（y_val）に分ける
X_val = df_val[['weekday', 'High', 'Low', 'Open', 'Close', 'Close_ratio', 'Body']]
y_val = df_val[['Up']]

# 検証データの説明変数と目的変数を確認
print(X_val)
print(y_val)
```

# データを整える

```python
# 標準化関数（StandardScaler）のインポート
from sklearn.preprocessing import StandardScaler

# numpyのインポート
import numpy as np

# 4日ごとにデータを抜き出して、標準化とnumpy配列に変換する関数（std_to_np）の定義
def std_to_np(df):
    df_list = []
    df = np.array(df)
    for i in range(0, len(df)-3, 4):
        df_s = df[i:i+4]
        scl = StandardScaler()
        df_std = scl.fit_transform(df_s)
        df_list.append(df_std)
    return np.array(df_list)
```

```python
# 学習データと検証データの説明変数に関数（std_to_np）を実行
X_train_np_array = std_to_np(X_train)
X_val_np_array = std_to_np(X_val)

# 学習データと検証データの形の確認
print(X_train_np_array.shape)
print(X_val_np_array.shape)
```

```python
# 学習データと検証データの目的変数を確認
print(y_train)
print(y_val)
```

```python
# 学習データ、検証データの目的変数の間引き
# 週の4日目（木曜日）のデータだけ抜き出す
y_train_new = y_train[3::4]
y_val_new = y_val[3::4]

# 間引き後の学習データと検証データの目的変数を確認
print(len(y_train_new))
print(len(y_val_new))
```

# 予測モデルを作成する

```python
# keras.modelsからSequentialのインポート
from keras.models import Sequential
# keras.leyersからDense, LSTMのインポート
from keras.layers import Dense, LSTM
# keras.leyersからDropoutのインポート
from keras.layers import Dropout
```

```python
# LSTM構築とコンパイル関数
def lstm_comp(df):
    # 入力層／中間層／出力層のネットワークを構築
    model = Sequential()
    model.add(LSTM(256, activation='relu', batch_input_shape=(None, df.shape[1], df.shape[2])))
    model.add(Dropout(0.2))
    model.add(Dense(256, actication='relu'))
    model.add(Dropout(0.2))
    model.add(Dense(1, actication='sigmoid'))

    # ネットワークのコンパイル
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model
```

# 交差検証で予測精度を確認する

```python
# 時系列分割のためTimeSeriesSplitのインポート
from sklearn.model_selection import TimeSeriesSplit
# accuracy算出のためaccuracy_scoreのインポート
from sklearn.metrics import accuracy_score
```

```python
valid_scores = []
tscv = TimeSeriesSplit(n_splits=4)

for fold, (train_indices, valid_indices) in enumerate(tscv.split(X_train_np_array)):
    X_train, X_valid = X_train_np_array[train_indices], X_train_np_array[valid_indices]
    y_train, y_valid = y_train_new[train_indices], y_train_new[valid_indices]

    # LSTM構築とコンパイル関数にX_trainに私、変数modelに代入
    model = lstm_comp(X_train)

    # モデル学習
    model.fit(X_train, y_train, epochs=10, batch_size=64)

    # 予測
    y_valid_pred = model.predict(X_valid)

    # 予測結果の2値化
    y_valid_pred = np.where(y_valid_pred < 0.5, 0, 1)

    # 予測精度の算出と表示
    score = accuracy_score(y_valid, y_valid_pred)
    print(f'fold {fold} MAE: {score}')

    # 予測精度スコアをリストに格納
    valid_scores.append(score)

print(f'valid_sccores: {valid_scores}')
cv_score = np.mean(valid_scores)
print(f'CV score: {cv_score}')
```
