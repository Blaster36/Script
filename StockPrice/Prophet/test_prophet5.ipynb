```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from pandas_datareader.data import DataReader
import datetime
```

```python
start = datetime.date(2018, 1, 1)
end = datetime.date(2022, 1, 1)

df = DataReader('9983.T', 'yahoo', start, end)
```

```python
df.info()
```

```python
df.head()
```

```python
df['ds'] = df.index
df = df.rename({'Close':'y'}, axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
# 不要カラムの削除と並べ替え
df = df[['ds', 'y']]
```

```python
df.info()
```

```python
df.head()
```

```python
# 学習データとテストデータの分割
df_train = df.loc[:'2021-01-01']
df_test = df.loc['2021-01-01':]
```

```python
df_train.info()
```

```python
df_train.head()
```

```python
df_train.tail()
```

```python
df_test.info()
```

```python
df_test.head()
```

```python
df_test.tail()
```

```python
params = {'growth': 'linear',
          'changepoints': None,
          'n_changepoints': 25,
          'changepoint_range': 0.8,
          'yearly_seasonality': 'auto',
          'weekly_seasonality': 'auto',
          'daily_seasonality': 'auto',
          'holidays': None,
          'seasonality_mode': 'additive',
          'seasonality_prior_scale': 10.0,
          'holidays_prior_scale': 10.0,
          'changepoint_prior_scale': 0.05,
          'mcmc_samples': 0,
          'interval_width': 0.80,
          'uncertainty_samples': 1000,
          'stan_backend': None
         }
model = Prophet(**params)

# Prophet 予測モデル構築
df_prophet_model = Prophet(**params)
df_prophet_model.fit(df_train)
```

```python
len(df_test)
```

```python
# Prophet 予測モデルの精度検証用データの生成
df_future = df_prophet_model.make_future_dataframe(periods=len(df_test), freq='d')
df_pred = df_prophet_model.predict(df_future)
```

```python
df_future.info()
```

```python
df_pred.info()
```

```python
df_pred['yhat']
```

```python
type(df_pred['yhat'])
```

```python
# Prophet 予測モデルの予測結果（学習データ期間＋テストデータ期間）
df_pred_plot = df_prophet_model.plot(df_pred)         #予想値（点は学習データの実測値）
df_pmpc = df_prophet_model.plot_components(df_pred)   #モデルの要素分解
```

```python
# テストデータに予測値を結合
df_test.loc[:, 'Prophet Predict'] = df_pred['yhat'][-245:].to_list()
```

```python
df_test.info()
```

```python
plt.figure(figsize=[20, 15])

# 実測データと予測データの折れ線グラフを描画
df_test['y'].plot(title='Forecast evaluation', kind='line')
df_test['Prophet Predict'].plot(title='Forecast evaluation', kind='line')

# グラフの凡例を設定
plt.legend(['y', 'Prophet Predict'])

# x軸の範囲を変更する
#plt.xlim('2021-10-01', '2021-12-31')
# y軸の範囲を変更する
plt.ylim(60000, 140000)

# グラフの表示
plt.show()
```
