```python
import pandas as pd
```

```python
# CSVファイルからDataFrameを生成
df = pd.read_csv('livedoor_news/livedoor_news_straight.csv', encoding='utf-8')
df.head()
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("/home/jovyan/bert-base-japanese-sentiment")
tokenizer = AutoTokenizer.from_pretrained("/home/jovyan/bert-base-japanese-whole-word-masking", word_tokenizer_type="mecab", mecab_kwargs={"mecab_option": "-d /home/jovyan/mecab/ipadic -r /dev/null"})

sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
for title in df['title'].to_list()[:5]:
    print(sentiment_analyzer(title), title)
```

```python
# DataFrameに列を追加し、各ニュースにラベルとスコアを割り振る
df['label'] = df['title'].apply(lambda x: sentiment_analyzer(x)[0]['label'])
df['score'] = df['title'].apply(lambda x: sentiment_analyzer(x)[0]['score'])
```
